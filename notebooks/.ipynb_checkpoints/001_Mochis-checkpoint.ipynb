```python
import enerhabitat as eh
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
eh.materials('materials.ini')
eh.get_list_materials()
```

```python
eh.Nx = 100
eh.dt = 60

dia = eh.Tsa(
    epw_file='../epw/MEX_MOR_Zacatepec.767270_TMYx.2007-2021.epw',
    solar_absortance=0.3,
    surface_tilt=90,
    surface_azimuth=270,
    month='02'
)
sc = [
    # (0.254,"poliestirenoestandard"),
    # (0.1, "adobe"),
    # (0.1, "adobe"),
#     (0.02, "tabique"),
     (0.05, "concreto"),
     (0.05, "concreto"),
]

dia = eh.solveCS(
    sc,
    dia
)

fig, ax = plt.subplots(figsize=(12,3))

ax.plot(dia.Ta, label='Ta')
ax.plot(dia.Tsa, label='Tsa')
ax.plot(dia.Ti, label='Ti')
ax.legend()
```

```python
df = dia.reset_index()

px.line(
    data_frame=df,
    x='index',
    y=['Tsa','Ti','Ta'],
)
```
