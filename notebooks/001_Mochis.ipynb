```python
import enerhabitat as eh
import matplotlib.pyplot as plt
```

```python
eh.get_list_materials()
```

```python
eh.Nx = 100
eh.dt = 600

df = eh.Tsa(
    epw_file='../epw/MEX_MOR_Zacatepec.767270_TMYx.2007-2021.epw',
    solar_absortance=0.3, #color 0.3 blanco, 0.6 rojo, 0.9 negro
    surface_tilt=90, # 0 techo, 90 muro vertical
    surface_azimuth=270, # 0 norte, 90 este, 180 sur, 270 oeste
    month='02'
)
sc = [
     (0.15, "concreto"),
]

df = eh.solveCS(
    sc,
    df
)

fig, ax = plt.subplots(figsize=(12,3))

ax.plot(df.Ta, label='Ta')
ax.plot(df.Tsa, label='Tsa')
ax.plot(df.Ti, label='Ti')
ax.grid()
ax.legend()

concreto_blanco = df.copy()
```
